### Desafio 4:
Ordenar o dataframe para que a primeira linha seja a com maior gasto e adicionar uma coluna com a região de cada estado

```python
df_usaveis.sort_values(by='Total', ascending=False, inplace=True)
```

```python
def map_region(estado):
    return 'Sudeste' if estado in ['35 São Paulo', '31 Minas Gerais', '33 Rio de Janeiro', '32 Espírito Santo'] else \
              'Sul' if estado in ['41 Paraná', '43 Rio Grande do Sul', '42 Santa Catarina'] else \
                'Centro-Oeste' if estado in ['52 Goiás', '53 Distrito Federal', '50 Mato Grosso do Sul', '51 Mato Grosso'] else \
                    'Norte' if estado in ['11 Rondônia', '12 Acre', '13 Amazonas', '14 Roraima', '15 Pará', '16 Amapá', '17 Tocantins'] else \
                        'Nordeste'
df_usaveis['Regiao'] = df_usaveis.index.map(map_region)
```

```python
def plota_gastos_por_mes(df, figsize=(10,6)):
    ax = df.T.plot(figsize=figsize)

    ax.set_xlabel('Ano/Mês')
    ax.set_ylabel('Valor total em reais (milhões)')

    ax.title.set_text('Atendimentos por ano/mês')
    ax.legend(loc='best', title='Unidades da Federação')

    ax.plot()
```

```python
df_aux = df_usaveis.iloc[:,47:-2]/1_000_000
plota_gastos_por_mes(df_aux.head(5))
plt.ylim(0, 650)
```

### Desafio 6:
Passar uma linha horizontal no estado de referência, colorir o estado referência e os outros com tons diferentes dependendo se for maior ou menor.

```python
tabela_comparacao = df_recente / df_recente.loc['26 Pernambuco']
tabela_comparacao = tabela_comparacao.sort_values(ascending=False)
```

```python
colors = ['red' if x > 1 else 'blue' if x < 1 else 'gray' for x in tabela_comparacao]

ax = tabela_comparacao.plot(kind='bar', figsize=(10,5), color=colors, label='')
ax.axhline(y=1, color='gray', linestyle='--', label='Valor de referência (Pernambuco)')

ax.set_xlabel('Unidade da Federação')
ax.set_ylabel('Proporção em 2023/Mai')

ax.title.set_text('Proporção de gasto dos estados em relação a Pernambuco')
ax.legend(loc='best', title='Legenda')

ax.plot()
```
