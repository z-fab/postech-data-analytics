# Usando Prophet

A biblioteca Prophet foi desenvolvida pelo Facebook para prever séries temporais com base em um modelo aditivo em que tendências não lineares são ajustadas a sazonalidade anual, semanal e diária, além de efeitos de feriados. Funciona melhor com séries temporais que têm fortes efeitos sazonais e várias temporadas de dados históricos.

Prophet é robusto para a falta de dados e mudanças na tendência, e geralmente lida bem com outliers.

```python
import pandas as pd
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
```

```python
df = pd.read_csv(
    'https://raw.githubusercontent.com/facebook/prophet/main/examples/example_wp_log_peyton_manning.csv'
)
df.head()
```

```python
m = Prophet()
m.fit(df)
```

```python
df_resp = m.predict(m.make_future_dataframe(periods=365))
```

```python
plot_plotly(m, df_resp)
```
